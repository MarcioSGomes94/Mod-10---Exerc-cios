````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv(r"C:\Users\marci\Desktop\Márcio\Programação\EBAC\02 - Phyton\4 - Desenvolvimento de aplicações com Python\Mod 10\previsao_de_renda.csv")

df[['posse_de_veiculo', 'posse_de_imovel','renda']]
````

````python
df['log_renda'] = np.log(df['renda'])

# Gráfico: Posse de veículo vs. Log da Renda
plt.figure(figsize=(10, 5))
ax = sns.pointplot(x="posse_de_veiculo", y="log_renda", data=df, dodge=True, errorbar=('ci', 95))
plt.title("Posse de Veículo x Log da Renda")
plt.show()
````

````python
plt.figure(figsize=(10, 5))
ax = sns.pointplot(x="posse_de_imovel", y="log_renda", data=df, dodge=True, errorbar=('ci', 95))
plt.title("Posse de Imóvel x Log da Renda")
plt.show()
````

Analisando podemos notar que as pessoas com Imovel ou Carro tendem a ter uma renda maior sim, é uma boa percussora sim. A posso de Imovel retorna um log com intervalo menor do que o de renda, o que me leva a acreditar sem um indicador mais confiavel.
